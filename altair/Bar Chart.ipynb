# Bar Chart

## Build the basic chart

```python
import pandas as pd

df = pd.read_csv('data/meals.csv')
df.head()
```

```python
import altair as alt

chart = alt.Chart(df).mark_bar(
).encode(
    y=alt.Y('Meal Type'),
    x=alt.X('Number of Likes')
)

chart
```

```python
chart = alt.Chart(df).mark_bar(
).encode(
    y=alt.Y('Meal Type', sort='-x'),
    x=alt.X('Number of Likes')
)

chart
```

```python
chart = alt.Chart(df).mark_bar(
).encode(
    y=alt.Y('Meal Type', sort='-x', title=None),
    x=alt.X('Number of Likes', axis=None)
)

chart
```

```python
chart = alt.Chart(df).mark_bar(
    color='#81c01e'
).encode(
    y=alt.Y('Meal Type', sort='-x', title=None),
    x=alt.X('Number of Likes', axis=None)
)

chart
```

```python
chart = alt.Chart(df).mark_bar(
    color='#81c01e'
).encode(
    y=alt.Y('Meal Type', sort='-x', title=None),
    x=alt.X('Number of Likes',axis=None)
).properties(
    width=300,
    height=300,
    title='Number of Likes for Each Meal'
)

chart
```

## Add text at the end of each bar

```python
text = chart.mark_text(
).encode(
    text='Number of Likes',
    y=alt.Y('Meal Type', sort='-x', title=None)
)

chart + text
```

```python
text = chart.mark_text(
    align='left',
    baseline='middle',
).encode(
    text='Number of Likes',
    y=alt.Y('Meal Type', sort='-x', title=None)
)

chart + text
```

```python
text = chart.mark_text(
    align='left',
    baseline='middle',
    dx = 3
).encode(
    text='Number of Likes',
    y=alt.Y('Meal Type', sort='-x', title=None)
)

chart + text
```

```python
text = chart.mark_text(
    align='left',
    baseline='middle',
    dx = 3,
    fontSize = 14
).encode(
    text='Number of Likes',
    y=alt.Y('Meal Type', sort='-x', title=None)
)

chart + text
```

```python
text = chart.mark_text(
    align='left',
    baseline='middle',
    dx=3,
    color='#81c01e',
    fontSize=14
).encode(
    text='Number of Likes',
    y=alt.Y('Meal Type', sort='-x', title=None)
)

chart + text
```

## Add five vertical lines to the chart

```python
lines_df = pd.DataFrame({'x': [20, 40, 60, 80, 100]})
lines_df
```

```python
lines = alt.Chart(lines_df).mark_rule(
    color='white',
    strokeWidth=1,
    opacity=0.5
).encode(
    x='x:Q'
)

chart + text + lines
```

```python
final_chart = (chart + text + lines
).configure_view(
    strokeWidth=0
)

final_chart
```

```python
chart.save('chart.html')
```
